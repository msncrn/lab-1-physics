```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
preamp_data = pd.read_excel('lockinfinal.xlsx', sheet_name='Preamplifier', index_col=0)
```

```python
lowgain = preamp_data.drop(columns=['200 gain','500 gain','1000 gain'])
```

```python
plt.figure(figsize=(12,6))
plt.plot(lowgain['1 gain'], c='b')
plt.plot(lowgain['2 gain'], c='black')
plt.plot(lowgain['5 gain'], c='red')
plt.plot(lowgain['10 gain'], c='green')
plt.plot(lowgain['20 gain'], c='orange')
plt.plot(lowgain['50 gain'], c='y')
plt.plot(lowgain['100 gain'], c='dodgerblue')
plt.xlabel("Frequency")
plt.ylabel("Voltage Gain")
plt.yscale("log")
plt.legend(["1V", "2V", "5V", "10V", "20V", "50V", "100V"], loc='lower left')
plt.title("Low Gain Data")
plt.show()
```

```python
hgain = preamp_data.drop(columns=['1 gain','2 gain','5 gain', '10 gain', '20 gain', '50 gain', '100 gain'])
```

```python
plt.figure(figsize=(12,6))
plt.plot(hgain['200 gain'], c='b')
plt.plot(hgain['500 gain'], c='black')
plt.plot(hgain['1000 gain'], c='red')
plt.xlabel("Frequency")
plt.ylabel("Voltage Gain")
plt.yscale("log")
plt.legend(["200V", "500V", "1000V"], loc='lower left')
plt.title("High Gain Data")
plt.show()
```
